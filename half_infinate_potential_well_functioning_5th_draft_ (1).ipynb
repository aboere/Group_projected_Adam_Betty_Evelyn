##**The Half-Infinite Potential Well in Atomic Units**

In this notebook, we simulate the wavefunctions of a quantum particle in a half-infinite potential well, characterized by an impenetrable wall at $(0<x<L)$, a free region $(0<x<L)$, and a finite potential barrier $V=U$ for $x>L$. We solve the time-independent Schrödinger equation in each region and apply boundary conditions to find allowed energy levels and corresponding wavefunctions.

We solve the **half-infinite potential well** with the potential:

- \$(V = \infty\) for \(x < 0\)$ (impenetrable wall),
- \(V = 0\) for \(0 < x < L\) (free motion inside the well),
- \(V = U\) for \(x > L\) (finite potential barrier),

using the wavefunction in **Region III** as:

$$
\psi(x) = G e^{-\alpha x}, \quad \text{where} \quad \alpha = \sqrt{2(U - E)}
$$

---

## **Region 1**

The time-independent Schrödinger equation (TISE) is:

$$
-\frac{1}{2} \frac{d^2 \psi}{dx^2} + V(x) \psi = E \psi
$$

Since the potential is infinite for $(x<0)$, the wavefunction must vanish in this region :

$$
\psi(x) = 0, \quad x < 0
$$

Thus, the boundary condition at \$(x = 0\) $is:

$$
\psi(0) = 0
$$

---
## **Region 2**
In region 2 $(0<x<L)$, the potential is zero $(V=0)$, so the Time-independant Shrödinger equation simplifies to:

$$
\frac{d^2 \psi}{dx^2} + k^2 \psi = 0, \quad k = \sqrt{2E}
$$

The general solution is:

$$
\psi(x) = A \sin(kx) + B \cos(kx)
$$
<br>
Applying the boundary condition $\psi(0) = 0$, we find that $B=0$, leaving the simplified solution

$$
\psi(x) = A \sin(kx)
$$

---
## **Region 3**
In region 3 $(x>L)$, the potential is constant and equal to $V=U$, so the TISE becomes

$$
\frac{d^2 \psi}{dx^2} + \alpha^2 \psi = 0, \quad \alpha = \sqrt{2(U - E)}
$$

The general solution to this differential equation is:

$$
\psi(x) = F e^{\alpha x} + G e^{-\alpha x}
$$

Since a physically valid bound-state wavefunction must **decay at infinity**, we set \(F = 0\), leaving:

$$
\psi(x) = G e^{-\alpha x}
$$

---

## **Applying the Boundary Conditions**
To ensure a physically valid and smooth wavefunction, we apply bounday conditions at $X=L$, where regions 2 and 3 meet.
<br>
<br>
First, the wavefunction must be continuous at $X=L$:


$$
\psi(L^-) = \psi(L^+)
$$

Substituting the solutions from Regions 2 and 3:

$$
A \sin(kL) = G e^{-\alpha L}
$$

Next, the first derivative of the wavefunction must also be continuous at $X=L$:

$$
\frac{d\psi}{dx} \Big|_{x=L^-} = \frac{d\psi}{dx} \Big|_{x=L^+}
$$

Differentiating both sides:

$$
k A \cos(kL) = -\alpha G e^{-\alpha L}
$$

Using \$(G = A \sin(kL) e^{\alpha L}\)$, from the continuity condition, we substitute:

$$
k \cot(kL) = -\alpha
$$

---

## **Energy Quantization Condition**

By applying the boundary conditions at $x=L$ and eliminating the unknown constants, we derived a relationship between the wave-vector $k=2E$ and the decay constant $α=2(U−E)$. This leads to the transcendental equation:

$$
\sqrt{2E} \cot\left(\sqrt{2E} L \right) = -\sqrt{2(U - E)}
$$
This equation determines the allowed energy levels E for a particle in the half-infinite potential well and must be solved numerically, as it cannot be rearranged analytically.

---

## **Summary**

- The wavefunction inside the well is:

  $$
  \psi(x) = A \sin(kx)
  $$

- The wavefunction outside the well (\(x > L\)) decays exponentially:

  $$
  \psi(x) = G e^{-\alpha x}
  $$

- The energy levels are determined by solving:

  $$
  \sqrt{2E} \cot\left(\sqrt{2E} L \right) = -\sqrt{2(U - E)}
  $$

This result is in **atomic units** where \$(\hbar = 1, m = 1, e = 1\)$

---




```python
###############################
#Defines U and L and plots the well
################################



import numpy as np
import matplotlib.pyplot as plt

# Define the potential function
def potential_well(x, L, U):
    """Returns the potential V(x) for a half-infinite potential well."""
    if x < 0:
        return np.inf  # Infinite potential for x < 0
    elif 0 <= x <= L:
        return 0  # Zero potential for 0 < x < L
    else:
        return U  # Finite potential U for x > L

# Parameters
L = 2  # Length of the box
U = 100  # Finite potential for x > L

# Create an array of x values
x_values = np.linspace(-1, L + 2, 1000)  # Range of x values to plot

# Compute the potential for each x
V_values = [potential_well(x, L, U) for x in x_values]

# Plot the potential
plt.figure(figsize=(8, 6))
plt.plot(x_values, V_values, label=f"$V(x)$: $L = {L}$, $U = {U}$", color="blue", linewidth=2)

# Add labels and title
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$V(x)$", fontsize=14)
plt.title("Half-Infinite Potential Well", fontsize=16)

# Add annotations for the regions
plt.text(-0.5, U / 2, "$V(x) = \infty$", fontsize=12, color="red", verticalalignment="center")
plt.text(L / 2, -0.5, "$V(x) = 0$", fontsize=12, color="green", verticalalignment="center", horizontalalignment="center")
plt.text(L + 0.5, U / 2, f"$V(x) = {U}$", fontsize=12, color="purple", verticalalignment="center")

# Add vertical lines at x = 0 and x = L
plt.axvline(x=0, color="blue", linestyle="-", linewidth=2)
plt.axvline(x=L, color="black", linestyle="--", linewidth=1)

# Set y-axis limits to show the infinite potential
plt.ylim(0, U + 2)

# Add a legend
plt.legend(fontsize=12)

# Show the plot
plt.grid(alpha=0.5)
plt.show()
```

```python
#################################
#sloves the energy quntization eqation
#######################################3


import numpy as np
from scipy.optimize import root_scalar

# Define the function f(y) = 0
def f(y, U, L):
    return np.sqrt(2 * y) * (1 / np.tan(np.sqrt(2 * y) * L)) + np.sqrt(2 * (U - y))



# Range to search for solutions
y_min = 0.01  # Avoid y = 0 due to division by zero
y_max = U    # Upper bound (since sqrt(2*(U-y)) must be real)

# Step size for initial grid search
step_size = 0.0001

# Tolerance for root finding
tol = 1e-6

# Store solutions
solutions = []

# Grid search to find initial guesses
y_values = np.arange(y_min, y_max, step_size)
for y_guess in y_values:
    try:
        # Use root_scalar to find the root near y_guess
        result = root_scalar(f, args=(U, L), bracket=(y_guess, y_guess + step_size), method='brentq', xtol=tol)
        if result.converged:
            root = result.root
            # Check if the root is already found (to avoid duplicates)
            if not any(np.isclose(root, sol, atol=tol) for sol in solutions):
                solutions.append(root)
    except ValueError:
        # Skip intervals where the function is not valid (e.g., division by zero)
        continue

# Print all solutions
print("Numerical solutions With numerical artifacts:", solutions)

### Removes every other solution
### Every other solution is a mistake of the numerical solution. Because the function has asymptotes the
#numerical approach is simply assuming there's a massive change in the function and yielding a solution when in fact there's an asymptote
j=0
i=0
len_solotions_g = int(np.ceil(len(solutions)/2))
solutions_g = [0] * (len_solotions_g)


while j < len(solutions):
  solutions_g[i] = solutions[j]
  i+=1
  j+=2

print("Numerical solutions Without numerical artifacts:", solutions_g)

import matplotlib.pyplot as plt

y_values = np.linspace(y_min, y_max, 100)
plt.plot(y_values, f(y_values, U, L))
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('E')
plt.ylabel('Output f(E)')
plt.title('Plot of f(y)')
plt.show()
```



## **Normalizing**

The wavefunction must be normalized so that the total probability of finding the particle anywhere in space is 1:


$$
\int_{-\infty}^{\infty} |\psi(x)|^2 dx = 1
$$

Since the wavefunction \$( \psi(x) \)$ is zero in **Region I** (x < 0 \), we only integrate over **Region II** ( 0 < x < L \) and **Region III** ( x > L ):

$$
\int_0^L |\psi_{\text{II}}(x)|^2 dx + \int_L^\infty |\psi_{\text{III}}(x)|^2 dx = 1
$$

### **Wavefunction Expressions**
For **Region II** (\( 0 < x < L \), inside the well):

$$
\psi_{\text{II}}(x) = A \sin(kx)
$$

For **Region III** (\( x > L \), outside the well):

$$
\psi_{\text{III}}(x) = G e^{-\alpha x}
$$

### **Compute the Integral Over Region II**

$$
\int_0^L |\psi_{\text{II}}(x)|^2 dx = \int_0^L A^2 \sin^2(kx) dx
$$




### **Compute the Integral Over Region III**
We compute:

$$
\int_L^\infty |\psi_{\text{III}}(x)|^2 dx = \int_L^\infty G^2 e^{-2\alpha x} dx
$$

Therefore:
$$
\int_0^L |\psi_{\text{II}}(x)|^2 dx + \int_L^\infty |\psi_{\text{III}}(x)|^2 dx = \int_0^L A^2 \sin^2(kx) dx+\int_L^\infty G^2 e^{-2\alpha x} dx = 1
$$
### **Solve for Normalization Constant A**
$$
\psi(L^-) = \psi(L^+)
$$

Substituting the solutions,

$$
A \sin(kL) = G e^{-\alpha L}
$$

$$
A \sin(kL)e^{\alpha L} = G
$$
Substitute the expression for G into the normalization condition

$$
 \int_0^L A^2 \sin^2(kx) dx+G^2\int_L^\infty  e^{-2\alpha x} dx = 1
$$

$$
 \int_0^L A^2 \sin^2(kx) dx+(A \sin(kL)e^{\alpha L})^2\int_L^\infty e^{-2\alpha x} dx = 1
$$

$$
 \int_0^L A^2 \sin^2(kx) dx+A^2 \sin^2(kL)e^{\alpha L}\int_L^\infty e^{-2\alpha x} dx = 1
$$

$$
A^2=\frac{1}{\int_0^L  \sin^2(kx) dx+ \sin^2(kL)e^{\alpha L}\int_L^\infty e^{-2\alpha x} dx}
$$
Evaluating the integral leads to

$$
A = \frac{1}{\sqrt{\frac{L}{2} - \frac{\sin(2kL)}{4k} + \frac{\sin^2(kL)}{2\alpha}}}
$$

Substitute A into the expression for G

$$
G = A \sin(kL) e^{\alpha L}
$$

This fully determines constant A and G, completing the normalized wavefunction.

### **Final Expressions for \( A \) and \( G \)**
$$
A = \frac{1}{\sqrt{\frac{L}{2} - \frac{\sin(2kL)}{4k} + \frac{\sin^2(kL)}{2\alpha}}}
$$

$$
G = A \sin(kL) e^{\alpha L}=\frac{\sin(kL) e^{\alpha L}}{\sqrt{\frac{L}{2} - \frac{\sin(2kL)}{4k} + \frac{\sin^2(kL)}{2\alpha}}}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

#####Uses the derived expressions to determine A and G

def psi_II(x, A, k):
    return A * np.sin(k * x)

def psi_III(x, G, alpha):
    return G * np.exp(-alpha * x)

def normalize_wave_function(E):
    k = np.sqrt(2 * E)
    alpha = np.sqrt(2 * (U - E))


    A= 1 / np.sqrt(L / 2 - np.sin(2 * k * L) / (4 * k) + np.sin(k * L) ** 2 / (2 * alpha))
    G = A * np.sin(k * L) * np.exp(alpha * L)


    return A, G, k, alpha

##### All the code after this point is simply defining the plot

# Define the potential function for plotting
def potential_well(x, L, U):
    if x < 0:
        return np.inf  # Infinite potential for x < 0
    elif 0 <= x <= L:
        return 0  # Zero potential for 0 < x < L
    else:
        return U  # Finite potential U for x > L

# Create an array of x values for the potential plot
x_values_potential = np.linspace(-1, L + 2, 1000)
V_values = [potential_well(x, L, U) for x in x_values_potential]

# Plot the potential well
plt.figure(figsize=(12, 8))
plt.plot(x_values_potential, V_values, label=f"$V(x)$: $L = {L}$, $U = {U}$", color="blue", linewidth=2)

# Add annotations for the regions
plt.text(-0.5, U / 2, "$V(x) = \\infty$", fontsize=12, color="red", verticalalignment="center")
plt.text(L / 2, -0.5, "$V(x) = 0$", fontsize=12, color="green", verticalalignment="center", horizontalalignment="center")
plt.text(L + 0.5, U / 2, f"$V(x) = {U}$", fontsize=12, color="purple", verticalalignment="center")

# Add vertical lines at x = 0 and x = L
plt.axvline(x=0, color="blue", linestyle="-", linewidth=2)
plt.axvline(x=L, color="black", linestyle="--", linewidth=1)

# Set y-axis limits to show the infinite potential
plt.ylim(0, U + 2)

# Add a legend
plt.legend(fontsize=12)

# Plot the normalized wave functions
x_values_II = np.linspace(0, L, 1000)
x_values_III = np.linspace(L, L + 2, 1000)

for i, E in enumerate(solutions_g):
    A, G, k, alpha = normalize_wave_function(E)

    psi_II_values = psi_II(x_values_II, A, k)
    psi_III_values = psi_III(x_values_III, G, alpha)

    # Scale the wave functions for better visualization
    scale_factor = 2  # Adjust this factor to fit the wave functions on the plot
    plt.plot(x_values_II, psi_II_values * scale_factor + E, label=f'E_{i+1} = {E:.3f} (Region II)')
    plt.plot(x_values_III, psi_III_values * scale_factor + E, label=f'E_{i+1} = {E:.3f} (Region III)')

# Add labels and title
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$V(x)$ and $\\psi(x)$", fontsize=14)
plt.title("Half-Infinite Potential Well with Normalized Wave Functions", fontsize=16)

# Show the plot
plt.grid(alpha=0.5)
plt.show()
```
